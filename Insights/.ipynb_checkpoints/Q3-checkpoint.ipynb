# Analyzing budget impact on earnings



## Objective:

The goal is to analyzing the correlation between the production budget and gross earnings.

# Data:

the dataset is from kaggle, it's rich with information. the link: " https://www.kaggle.com/datasets/alessandrolobello/the-ultimate-film-statistics-dataset-for-ml?resource=download "

## First: Starting with the Data preprocessing

I will start by simply importing the necessary libraries and read the csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
Path="./data/movie_statistic_dataset.csv"
movie_data = pd.read_csv(Path)
```

I will start with replacing all the placeholder values like "-" and "\N" with NaN.

```python
movie_data = movie_data.replace(["-", "\\N"], pd.NA)
movie_data
```

There might be rows with missing production budget or worldwide gross value and they shall be dropped for having no purpose in this project.

```python
movie_data.dropna(subset=['Production budget $', 'Worldwide gross $'])
```

Then converting the columns data types to appropriate data types as a double check for errors.

```python
movie_data['Production budget $'] = pd.to_numeric(movie_data['Production budget $'])
movie_data['Worldwide gross $'] = pd.to_numeric(movie_data['Worldwide gross $'])

movie_data.info()
```

A simple check for the first rows as confirmation.

```python
movie_data.head()
```

## Secondly, Exploratory Data Analysis (EDA)

I will start by describing the numerical vlaues of both variables.

```python
movie_data[['Production budget $', 'Worldwide gross $']].describe()
```

### Visualizations:

Visualizing  the Distribution of Movies by Production Budget.

```python
bins = [0, 1e7, 5e7, 1e8, 2e8, 5e8] 
labels = ['< $10M', '$10M-$50M', '$50M-$100M', '$100M-$200M', '> $200M']


movie_data['Budget Category'] = pd.cut(movie_data['Production budget $'], bins=bins, labels=labels, include_lowest=True)


budget_distribution = movie_data['Budget Category'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(budget_distribution, labels=budget_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Movies by Production Budget')
plt.show()
```

Visualizing the Distribution of Movies by Worldwide Gross Earnings.

```python

gross_bins = [0, 1e7, 5e7, 1e8, 5e8, 1e9, 3e9]  # Gross ranges in dollars
gross_labels = ['< $10M', '$10M-$50M', '$50M-$100M', '$100M-$500M', '$500M-$1B', '> $1B']


movie_data['Gross Category'] = pd.cut(movie_data['Worldwide gross $'], bins=gross_bins, labels=gross_labels, include_lowest=True)


gross_distribution = movie_data['Gross Category'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(gross_distribution, labels=gross_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Distribution of Movies by Worldwide Gross Earnings')
plt.show()
```

A scatter plot showing the relation between production budget and worldwide gross earnings.

```python



plt.figure(figsize=(10, 6))
plt.scatter(movie_data['Production budget $'], movie_data['Worldwide gross $'], alpha=0.5)
plt.title('Production Budget vs Worldwide Gross Earnings')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.xscale('log')  
plt.yscale('log')
plt.show()
```

A plot showing the relation between production budget and worldwide gross earnings.

```python
plt.figure(figsize=(10, 6))
plt.plot(movie_data['Production budget $'], movie_data['Worldwide gross $'], alpha=0.5)
plt.title('Production Budget vs Worldwide Gross Earnings')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.xscale('log')  
plt.yscale('log')
plt.show()
```

### Correlation:

Calculating Spearman's correlation

```python


spearman_corr, p_value = spearmanr(movie_data['Production budget $'], movie_data['Worldwide gross $'])
print(f"Spearman's rank correlation: {spearman_corr:.2f}")
print(f"P-value: {p_value:.4f}")
```

#### Additional details:

Pivot_table to summarize the average production budget and worldwide gross earnings grouped by budget category

```python


budget_bins = [0, 1e7, 5e7, 1e8, 2e8, 5e10] 
budget_labels = ['<$10M', '$10M-$50M', '$50M-$100M', '$100M-$200M', '>$200M']

movie_data['Budget Category'] = pd.cut(movie_data['Production budget $'], bins=budget_bins, labels=budget_labels)


pivot_table_budget = movie_data.pivot_table(
    values=['Worldwide gross $', 'Production budget $','movie_title'],
    index='Budget Category',
    aggfunc={'Production budget $': 'mean', 'Worldwide gross $': 'mean', 'movie_title': 'count'}
)


print(pivot_table_budget)
```

## Finally, Results:

Depending on the result we can estimate the effect of the budget of the movie on the earnings and their correlation.
