# Analyzing the data

## Setting up for the analysis

#### Importing the neccessary libraries.

```python
import pandas as pd
```

#### Read the csv file.

```python
path = "./data/movie_statistic_dataset.csv"  
movie_data = pd.read_csv(path)
```

#### Display the information of the dataset.

```python
movie_data.info()
```

#### Display the first 5 rows from the dataset.

```python
movie_data.head()
```

#### Display the number of "\N".

```python
(movie_data == "\\N").sum()
```

#### Display the number of "-".

```python
(movie_data== "-").sum()
```

#### Checking for the uniqueness of the values.

```python
movie_data.nunique()
```

##### Checking for the number of duplicated movie_titles.

```python
movie_data['movie_title'].duplicated().sum()
```

#### Output the movie_dataset's numerical values such as count, min, max, mean.

```python
movie_data.describe()
```

#### Count the genre combinations.

```python
movie_data['genres'].value_counts()
```

#### Count individual genres.

```python
movie_data_genre = movie_data.drop('genres', axis=1).join(movie_data['genres'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('genres'))

movie_data_genre['genres'].value_counts()
```
